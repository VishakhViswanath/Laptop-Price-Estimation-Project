## Extracting the X Resolution and the Y Resolution

```python
# we will split the data into two sets at the resolution's "x" aka 'BY' notation

splitdf = ldf["ScreenResolution"].str.split('x', n=1, expand = True) #(if there is anything before x take that into considersation also)
splitdf
```

```python
ldf['x_res'] = splitdf[0]
ldf['y_res'] = splitdf[1]
ldf.head()
```

```python
"""
From the resolution columns we need to extract the digits only, 
we will use regex to find the exact matches we're looking for.
Snce the numbers are all scattered across the data, at first we will replace the 
"," with ""and then will find all numbers from the string as "\d+\.?\d+",

\d means integer number, 
\.? means all those numbers which come after the integer number,
\d+ means the string must end with an number
"""

ldf["x_res"] = ldf["x_res"].str.replace("," , " ").str.findall(r'(\d+\.?\d+)').apply(lambda x:x[0])

ldf.head()
```

```python
ldf['x_res'] = ldf["x_res"].astype('int')
ldf['y_res'] = ldf["y_res"].astype('int')

ldf.info()
```

```python

# Select only numerical columns
n_df = ldf.select_dtypes(include=['number'])

# Compute the correlation matrix
corr_matrix = n_df.corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap of Numerical Columns')
plt.show()
```

```python
n_df.corr()['Price']
```

From the correlated plot it is observed that laptop price increases as resolution increases which means x_res and y_res are positively correlated

Since x_res and y_res are having higher collinearity we can combine both to create a column with lesser collinearity named PPI(pixels per inch)

the formula for PPI is given below

$$
    PPI(pixels per inch) = \frac{\sqrt{X_resolution^2+Y_resolution^2}}{inches}
$$

```python
ldf["PPI"] = ((ldf["x_res"]**2 + ldf["y_res"]**2)**0.5 / ldf["Inches"]).astype('float')
ldf.head()
```

```python
n_df.corr()['Price']
```

Since most of the processors are made by intel and the major ones are i3,i5,i7 we will be clustering them into i3,i5,i7 & others, where others will be everything in the intel brand excluding the above three.

The second category will be AMD which itself is a different brand hence we will consider it as a different category itself.
We need to extract the first 3 words of the CPU column as the first 3 word describes the type of CPU

```python
ldf["CPU_name"] = ldf["Cpu"].apply(lambda text:" ".join(text.split()[:3]))
ldf.head()
```

```python
## Creating a function to classify the CPUs as Intel and Non Intel
```

```python
"""
Here we will create a function to classify the processors as i3, i5, 17 and 
if it belongs ito intel but not the above, we will categorise it as other intekl
processors, if not other non intel processors
"""

def processor_type(text):
    
    if text == "Intel Core i3" or text == "Intel Core i5" or text == "Intel Core i7":
        return text
    
    else:
        if text.split()[0] == "Intel":
            return "Other Intel processor"
        
        else:
            return "AMD Processor"
        
        
        
ldf["CPU_name"] = ldf["CPU_name"].apply(lambda text:processor_type(text))
ldf.head()
```

In memory column also we will split the type of memory and numerical value into two as we've done in the baiove case
Since we do not have the memory as a complete description as in`128GB SSD +  1TB HDD` we will perform certain modifications to the memory column to make it into the same dimension.

```python
ldf["Memory"].iloc[:1][0]
```

```python
# the four most commonly pobserved variations are: HDD, SSD, Flash, Hybrid


# to remove the decimal points
ldf["Memory"] = ldf["Memory"].astype(str).replace('\.0', '', regex = True)

# to replace "GB"
ldf["Memory"] = ldf["Memory"].str.replace("GB", " ")

# to replace "TB" with "000"
ldf["Memory"] = ldf["Memory"].str.replace("TB","000")

# Split the word across the '+' character
new_df = ldf["Memory"].str.split("+", n=1, expand = True)

new_df
```

```python
ldf["first"] = new_df[0]
ldf["first"] = ldf["first"].str.strip()
ldf.head()
```

```python
# a function to classify memoryto the 4 major types

def apply_changes(value):
    ldf['Layer_1' + value] = ldf['first'].apply(lambda x:1 if value in x else 0)

    
listtoapply = ["HDD", "SSD", "Hybrid", "FlashStorage"]

for value in listtoapply:
    apply_changes(value)
    
    
ldf.head()
```

```python
# Remove all the charcters from first except numerics

ldf["first"] = ldf["first"].str.replace(r'\D','', regex=True)
ldf["first"].value_counts()
```

```python
ldf['Second'] = new_df[1]
ldf.head()
```

```python
def apply_changes1(value):
    ldf['Layer_2' + value] = ldf['Second'].apply(lambda x:1 if value in x else 0)

    
listtoapply1 = ["HDD", "SSD", "Hybrid", "FlashStorage"]
ldf["Second"] = ldf["Second"].fillna("0")
for value in listtoapply1:
    apply_changes1(value)
    
    
# remove all characters except numbers

ldf["Second"] = ldf["Second"].str.replace(r'\D',"", regex =True)
ldf['Second'].value_counts()
```

```python
ldf["first"] = ldf["first"].astype('int')
ldf["Second"] = ldf["Second"].astype('int')
ldf.head()
```

```python
# Multiplying the elements and storing the result in subsequent columns

ldf["HDD"] = (ldf["first"]*ldf["Layer_1HDD"] + ldf["Second"]*ldf["Layer_2HDD"])
ldf["SSD"] = (ldf["first"]*ldf["Layer_1SSD"] + ldf["Second"]*ldf["Layer_2SSD"])
ldf["Hybrid"] = (ldf["first"]*ldf["Layer_1Hybrid"] + ldf["Second"]*ldf["Layer_2Hybrid"])
ldf["Flash_Storage"] = (ldf["first"]*ldf["Layer_1FlashStorage"] + ldf["Second"]*ldf["Layer_2FlashStorage"])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn import metrics
from sklearn.model_selection import RandomizedSearchCV
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
# from xgboost import XGBRegressor
# from sklearn.svm import SVR
# from sklearn.neighbors import KNeighborsRegressor
from sklearn import tree
```

```python
X_train, X_test, y_train, y_test = train_test_split(train,
                                                    test,
                                                    test_size=0.15,
                                                    random_state=42)
X_train.shape, X_test.shape
```

Since we are going to build Pipelines using column transformer class, for this we must get the index number of columns having categorical values

```python
map = { i:value for i, value in enumerate(X_train.columns)}
map
```

# 1. Linear Regression model

Since there are categorical columns in the df we will apply one hot encoding to columns 0,1,3,8,11

```python
step1 = ColumnTransformer(transformers=
                          [
    ('col_tnf', OneHotEncoder(sparse_output=False, drop ='first'),[0,1,3,8,11])
                        ], 
remainder = "passthrough")


step2 = LinearRegression()

pipe = Pipeline([
    ("step1", step1),
    ("step2", step2)
])

pipe.fit(X_train, y_train)

y_pred = pipe.predict(X_test)



print("R2 score", metrics.r2_score(y_test,y_pred))
print("MAE", metrics.mean_absolute_error(y_test,y_pred))


```

# 2. Ridge Regression model

```python
step1 = ColumnTransformer(transformers=
                          [
    ('col_tnf', OneHotEncoder(sparse_output=False, drop ='first'),[0,1,3,8,11])
                        ],remainder = "passthrough")


step2 = Ridge(alpha=10)

pipe = Pipeline([
    ("step1", step1),
    ("step2", step2)
                ])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print("R2 score", metrics.r2_score(y_test,y_pred))
print("MAE", metrics.mean_absolute_error(y_test,y_pred))
```

# 3. Lasso Regression
Here alsow we will apply one hot encoding on columns with indices >[0,1,3,8,11]

```python
step1 = ColumnTransformer(transformers=
                          [
    ('col_tnf', OneHotEncoder(sparse_output=False, drop ='first'),[0,1,3,8,11])
                        ],remainder = "passthrough")


step2 = Lasso(alpha=0.001)

pipe = Pipeline([
    ("step1", step1),
    ("step2", step2)
                ])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print("R2 score", metrics.r2_score(y_test,y_pred))
print("MAE", metrics.mean_absolute_error(y_test,y_pred))
```

# 4. Decision Tree

```python
step1 = ColumnTransformer(
transformers=[
    ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0,1,3,8,11])
], remainder='passthrough')

step2 = DecisionTreeRegressor(max_depth=10)

pipe = Pipeline([
    ('step1', step1),
    ('step2', step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)


print('R2 score', metrics.r2_score(y_test,y_pred))
print('MAE', metrics.mean_absolute_error(y_test,y_pred))
```

# 5. Random Forest

```python
step1 = ColumnTransformer(
transformers=[
    ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0,1,3,8,11])
], remainder='passthrough')

step2 = RandomForestRegressor(n_estimators=110,
                              random_state=42,
                              max_samples=0.5,
                              max_features=0.80,
                              max_depth=16)

pipe = Pipeline([
    ('step1', step1),
    ('step2', step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)


print('R2 score', metrics.r2_score(y_test,y_pred))
print('MAE', metrics.mean_absolute_error(y_test,y_pred))
```
